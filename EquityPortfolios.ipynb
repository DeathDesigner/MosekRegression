# Management of equity portfolios

We solve common problems in portfolio management. 

In all examples discussed below, the matrix $X \in \mathbb{R}^{n \times m}$ describes a time series of $n$ consecutive historic returns of $m$ risky assets, each row index of $X$ corresponding to an investment period, and each column index to an asset. We search for an *optimal* combination of those assets. Different objectives, expectations, constraints, risk appetite, and trading costs, etc. result in a variety of formulations used in practice.  

## Conventions


An investor wishes to manage a portfolio in $m$ risky
assets. The investor allocates fractions $w_i(t)$ of the risk capital $C$ in asset $i$ over an investment period $[t,t+1]$ (one hour, day, week, month, ...), at
the end of which he/she is prepared to adjust the positions again.

At time $t+1$ the update in $w$ is induced by changes in the underlying input data for the portfolio problem.

The vector $X w = (R_1,R_2,\ldots,R_n)^{T}$ describes the time series of $n$ consecutive portfolio returns. A portfolio return is the weighted sum of the $m$ linear returns

$$R_k = \sum_{i=1}^m w_i x_{k,i} = X_{k,\cdot} w$$

where $x_{k,i}$ is the return of asset $i$ over the $k$-th historic investment period.

In all experiments given below we compute $w$ and keep it constant throughout the test. In practice this enforces to adjust the positions frequently to keep the same relative fractions invested. Those adjustments induce trading costs. We ignore trading costs here and illustrate only the use of some of the tools introduced in the paper.

## The minimum variance portfolio

Although one could avoid being exposed to risk by not investing at all, minimum variance portfolios got popular. Such portfolios can be constructed by noting that they track a return time series with $r_M = (0, 0, \ldots, 0)$.
From a theoretical perspective, it should not make sense to invest in such portfolios, but in practice they are observed to perform quite competitively, hence their popularity.


$$
\begin{array}{llll} 
\mbox{minimize}& \left||X w\right||_2& &\\
\mbox{subject to}& \sum w_i = 1, \\ 
& w \geq 0&&
\end{array}
$$ 

```python
def MinimumVariance(X):
    # define model
    model = mModel.build_model('minVariance')
    
    # introduce n non-negative weight variables
    weights = mModel.weights(model, "weights", n=X.shape[1], lb=0.0)
    
    # e'*w = 1
    mBound.equal(model, Expr.sum(weights), 1.0)

    # minimization of the residual
    mModel.minimise(model, mMath.l2_norm(model, "2-norm(res)", mMath.mat_vec_prod(X.values, weights)))
    
    return pd.Series(data=np.array(weights.level()), index=X.keys())
```

```python
weights = MinimumVariance(stocks)
print weights
print weights.sum()
print weights.abs().sum()

profit = (stocks * weights).sum(axis=1)
(profit + 1.0).cumprod().plot()
```

## The mean-variance portfolio

The classical Markowitz portfolio optimization problem considers investing in $n$ stocks or assets held over a period of time. Let $w_j$ denote the amount invested in asset $j$, and assume a stochastic model where the return of the assets is a random variable $r$ with known mean $\mu$ and covariance $C$:

```python
mu = stocks.mean()
print mu
```

```python
covar = stocks.cov()
print covar
```

The problem facing the investor is to rebalance the portfolio to achieve a good compromise between risk and expected return.
We introduce a bound $L$ on the leverage of the portfolio. Using $L=1$ results in a classic long only portfolio whereas $L=1.6$ corresponds to a $130/30$ portfolio. This leads to the optimization problem

 
$$
\begin{array}{llll} 
\mbox{maximize}& {\mu}^{T} w - \alpha\; w^T C w& &\\
\mbox{subject to}& \sum w_i = 1, \\ 
& \sum \left|w_i\right| \leq L&&
\end{array}
$$ 

```python
def Markowitz(mu, covar, alpha=0.5, L=1.0):

    # define model
    model = mModel.build_model('meanVariance')
    
    # introduce n non-negative weight variables
    weights = mModel.weights(model, "weights", n=covar.shape[1])
    
    lev = mMath.l1_norm(model, "leverage", weights)
    # e'*w = 1
    mBound.equal(model, Expr.sum(weights), 1.0)
    25,
    mBound.upper(model, lev, L)
    
    v = Expr.dot(mu.values, weights)
    var = mMath.variance(model, "variance", weights, alpha*covar.values)
    
    mModel.maximise(model, Expr.sub(v,var))
    
    weights = pd.Series(data=np.array(weights.level()), index=stocks.keys())
    return weights
```

```python
weights = pd.DataFrame({alpha: Markowitz(mu, covar, alpha, L=4.0) for alpha in [0.1,0.25,0.5,0.75,1.0,1.25,1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0]})
print "Weights vs. alpha"
print weights

profit = pd.DataFrame({alpha: (stocks * weights[alpha]).sum(axis=1) for alpha in weights.columns})
print "Mean of Profits vs. alpha"
print profit.mean()
print "Stdev of Profits vs. alpha"
print profit.std()
print "Sharpe Ratio of Profits vs. alpha"
print profit.mean()/profit.std()

import matplotlib.pyplot as plt
plt.plot(profit.std(),profit.mean())
plt.title("Efficient Frontier")
```

## Adding trading costs

We do again the minimum variance portfolio but add the cost term $0.05\times\left|w-w_0\right|$ in the objective. 


```python
w0 = pd.Series(index=stocks.keys(), data=0.2)
print w0
```

```python
# define model
model = mModel.build_model('meanVariance')

# introduce n non-negative weight variables
weights = mModel.weights(model, "weights", n=stocks.shape[1], lb=0.0)

### this code fragment is still awkward as Mosek Fusion can't index an Expr at the moment (may change soon)
t = model.variable("trade", stocks.shape[1], Domain.unbounded())

# (t_i, w_i) \in Q2
for i in range(0, stocks.shape[1]):
    mMath.quad_cone(model, t.index(i), Expr.sub(weights.index(i),w0.values[i]))

### this is how the code should look like:
### t = mMath.absolute(model, "trade", Expr.sub(weights,w0))


# e'*w = 1
mBound.equal(model, Expr.sum(weights), 1.0)

v = mMath.l2_norm(model, "2-norm(res)", mMath.mat_vec_prod(stocks.values, weights))

cost = Expr.sum(t)

# minimization of the residual
mModel.minimise(model, Expr.add(v,Expr.mul(0.05,cost)))

weights = pd.Series(data=np.array(weights.level()), index=stocks.keys())
print weights
```

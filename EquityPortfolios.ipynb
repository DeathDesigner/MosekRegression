# Management of equity portfolios

We solve common problems in portfolio management. 

In all examples discussed below, the matrix $X \in \mathbb{R}^{n \times m}$ describes a time series of $n$ consecutive historic returns of $m$ risky assets, each row index of $X$ corresponding to an investment period, and each column index to an asset. We search for an *optimal* combination of those assets. Different objectives, expectations, constraints, risk appetite, and trading costs, etc. result in a variety of formulations used in practice.  

## Conventions


An investor wishes to manage a portfolio in $m$ risky
assets. The investor allocates fractions $w_i(t)$ of the risk capital $C$ in asset $i$ over an investment period $[t,t+1]$ (one hour, day, week, month, ...), at
the end of which he/she is prepared to adjust the positions again.

At time $t+1$ the update in $w$ is induced by changes in the underlying input data for the portfolio problem.

The vector $X w = (R_1,R_2,\ldots,R_n)^{T}$ describes the time series of $n$ consecutive portfolio returns. A portfolio return is the weighted sum of the $m$ linear returns

$$R_k = \sum_{i=1}^m w_i x_{k,i} = X_{k,\cdot} w$$

where $x_{k,i}$ is the return of asset $i$ over the $k$-th historic investment period.

In all experiments given below we compute $w$ and keep it constant throughout the test. In practice this enforces to adjust the positions frequently to keep the same relative fractions invested. Those adjustments induce trading costs. We ignore trading costs here and illustrate only the use of some of the tools introduced in the paper.

## Loading data
We have prepared some data and stored in csv file. We load this data and compute the returns. To deal with time series data we use [pandas](http://pandas.pydata.org/)

```python
import pandas as pd
pd.__version__
```

Function to compute the returns for the columns of a pandas dataframe:

```python
def returns(frame):
    def __compute_return(ts):
        ts = ts.dropna()
        return ts.diff() / ts.shift(1)

    return frame.apply(__compute_return).fillna(value=0.0)
```

Load data from a csv file and compute returns

```python
import os
data = pd.read_csv(os.path.join("data", "data.csv"), index_col=0, parse_dates=True)

stocks = returns(data[["GOOG", "T", "AAPL", "GS", "IBM"]])
```

Import numpy, Expr from mosek fusion and the MosekTools

```python
import numpy as np

from mosek.fusion import Expr

import mosekTools.solver.model.model as mModel
import mosekTools.solver.model.math as mMath
import mosekTools.solver.model.bound as mBound
```

```python
matrix = stocks.values
```

## The 1/N portfolio

This a long-only fully invested portfolio where each asset has is kept at same cash position. Hence the return at time $t$ is the mean of the individual stock returns. 

```python
    history = stocks.mean(axis=1)
    (history + 1.0).cumprod().plot()
```

## The minimum variance portfolio

Although one could avoid being exposed to risk by not investing at all, minimum variance portfolios got popular. Such portfolios can be constructed by noting that they track a return time series with $r_M = (0, 0, \ldots, 0)$.
From a theoretical perspective, it should not make sense to invest in such portfolios, but in practice they are observed to perform quite competitively, hence their popularity.


```python
# define model
model = mModel.build_model('minVariance')

# introduce n non-negative weight variables
weights = mModel.weights(model, "weights", n=matrix.shape[1], lb=0.0)

# e'*w = 1
mBound.equal(model, Expr.sum(weights), 1.0)

v = mMath.l2_norm(model, "2-norm(res)", expr=mMath.mat_vec_prod(matrix, weights))

# minimization of the residual
mModel.minimise(model=model, expr=v)

sol = np.array(weights.level())
```

```python
weights = pd.Series(index=stocks.keys(), data=sol)
print weights
print weights.sum()
```

```python
history = (stocks * weights).sum(axis=1)
(history + 1.0).cumprod().plot()
```
